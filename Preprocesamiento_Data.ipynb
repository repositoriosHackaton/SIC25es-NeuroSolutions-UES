```python
import pandas as pd
import matplotlib.pyplot as plt


ruta_archivo = "/kaggle/working/full_df.csv"
datos = pd.read_csv(ruta_archivo)


print(" Información general del dataset ")
print(datos.info())
print("\nNúmero total de registros y columnas:", datos.shape)


print("\n Valores nulos por columna ")
print(datos.isnull().sum())

print("\n Distribución de valores en 'diagnóstico' ")
print(datos['diagnostico'].value_counts())

etiqueta_cols = ['N', 'D', 'G', 'C', 'A', 'H', 'M', 'O']
print("\n Estadísticas descriptivas de las etiquetas binarias ")
estadisticas = datos[etiqueta_cols].describe()
print(estadisticas)

plt.figure(figsize=(12, 6))
estadisticas.loc['mean'].plot(kind='bar', color='purple', alpha=0.7)
plt.title("Media de las Etiquetas Binarias")
plt.xlabel("Etiquetas Binarias")
plt.ylabel("Media")
plt.xticks(rotation=0)
plt.show()

datos['id_original'] = datos['ruta'].apply(lambda x: x.split('_')[0])

duplicacion = datos.groupby('id_original').size()
print("\n Distribución de registros por 'id_original' ")
print(duplicacion.value_counts())

duplicacion_incompleta = duplicacion[duplicacion != 2]
if not duplicacion_incompleta.empty:
    print("\n>>> Alerta: Se encontraron IDs con duplicación incompleta:")
    print(duplicacion_incompleta)
else:
    print("\nTodos los 'id_original' tienen duplicación correcta (2 registros por ID).")

ids_inconsistentes = []
for id_val, grupo in datos.groupby('id_original'):
    if any(grupo[etiqueta_cols].nunique() > 1):
        ids_inconsistentes.append(id_val)

if ids_inconsistentes:
    print("\n>>> Alerta: Se encontraron 'id_original' con etiquetas inconsistentes:")
    print(ids_inconsistentes)
else:
    print("\nTodas las entradas duplicadas tienen etiquetas consistentes.")


print("\n Primeras 5 filas del dataset ")
print(datos.head())

print("\n Últimas 5 filas del dataset ")
print(datos.tail())
```

```python
import pandas as pd

ruta_datos = "/kaggle/working/data.xlsx"
ruta_full_df = "/kaggle/working/full_df.csv"

datos_excel = pd.read_excel(ruta_datos)
datos_csv = pd.read_csv(ruta_full_df)

print("\n=== Archivos cargados ===")
print(f"Data.xlsx: {datos_excel.shape}")
print(f"Full_df_processed.csv: {datos_csv.shape}")

duplicados_excel = datos_excel[datos_excel.duplicated(subset=['ruta'], keep=False)]
if not duplicados_excel.empty:
    print("\n=== Duplicados en Data.xlsx ===")
    print(duplicados_excel)
else:
    print("\n✔ No se encontraron duplicados en 'ruta' en Data.xlsx.")

duplicados_csv = datos_csv[datos_csv.duplicated(subset=['ruta'], keep=False)]
if not duplicados_csv.empty:
    print("\n=== Duplicados en Full_df_processed.csv ===")
    print(duplicados_csv)
else:
    print("\n✔ No se encontraron duplicados en 'ruta' en Full_df_processed.csv.")

rutas_multidiagnosticos_excel = datos_excel.groupby('ruta')['diagnostico'].nunique()
rutas_multidiagnosticos_excel = rutas_multidiagnosticos_excel[rutas_multidiagnosticos_excel > 1]
if not rutas_multidiagnosticos_excel.empty:
    print("\n=== Rutas con múltiples diagnósticos en Data.xlsx ===")
    print(rutas_multidiagnosticos_excel)
else:
    print("\n✔ Todas las rutas en Data.xlsx tienen un único diagnóstico.")

rutas_multidiagnosticos_csv = datos_csv.groupby('ruta')['diagnostico'].nunique()
rutas_multidiagnosticos_csv = rutas_multidiagnosticos_csv[rutas_multidiagnosticos_csv > 1]
if not rutas_multidiagnosticos_csv.empty:
    print("\n=== Rutas con múltiples diagnósticos en Full_df_processed.csv ===")
    print(rutas_multidiagnosticos_csv)
else:
    print("\n✔ Todas las rutas en Full_df_processed.csv tienen un único diagnóstico.")

datos_excel_limpios = datos_excel.drop_duplicates(subset=['ruta'], keep='first')
ruta_salida_excel = "/kaggle/working/data_cleaned.xlsx"
datos_excel_limpios.to_excel(ruta_salida_excel, index=False)
print(f"\nDataset limpio exportado como '{ruta_salida_excel}' (Data.xlsx).")

datos_csv_limpios = datos_csv.drop_duplicates(subset=['ruta'], keep='first')
ruta_salida_csv = "/kaggle/working/full_df_cleaned.csv"
datos_csv_limpios.to_csv(ruta_salida_csv, index=False)
print(f"\nDataset limpio exportado como '{ruta_salida_csv}' (Full_df_processed.csv).")

print("\n=== Resumen de datasets limpios ===")
print("Data.xlsx:")
print(f"Registros totales después de limpieza: {datos_excel_limpios.shape[0]}")
print(f"Columnas: {datos_excel_limpios.columns}")

print("\nFull_df_processed.csv:")
print(f"Registros totales después de limpieza: {datos_csv_limpios.shape[0]}")
print(f"Columnas: {datos_csv_limpios.columns}")

print("\nPrimeras filas del dataset limpio de Data.xlsx:")
print(datos_excel_limpios.head())

print("\nPrimeras filas del dataset limpio de Full_df_processed.csv:")
print(datos_csv_limpios.head())
```

```python
import pandas as pd
from sklearn.utils import resample

# Cargar datasets
ruta_datos = "/kaggle/working/data_cleaned.xlsx"
ruta_full_df = "/kaggle/working/full_df_cleaned.csv"

datos_excel = pd.read_excel(ruta_datos)
datos_csv = pd.read_csv(ruta_full_df)

# Agrupación de diagnósticos raros en Full_df_cleaned.csv
print("\nDistribución de diagnósticos")
distribucion_diagnosticos = datos_csv['diagnostico'].value_counts()
print(distribucion_diagnosticos)

umbral = 10
diagnosticos_poco_frecuentes = distribucion_diagnosticos[distribucion_diagnosticos < umbral]

print("\nDiagnósticos raros (frecuencia < 10)")
print(diagnosticos_poco_frecuentes)
print(f"\nTotal de diagnósticos raros: {len(diagnosticos_poco_frecuentes)}")
print(f"Porcentaje de diagnósticos raros: {(len(diagnosticos_poco_frecuentes) / len(distribucion_diagnosticos)) * 100:.2f}%")

datos_csv_agrupados = datos_csv.copy()
datos_csv_agrupados['diagnostico'] = datos_csv_agrupados['diagnostico'].apply(
    lambda x: "Otros" if x in diagnosticos_poco_frecuentes.index else x
)

print("\nDistribución de diagnósticos después de agrupar")
print(datos_csv_agrupados['diagnostico'].value_counts())

ruta_salida_grouped = "/kaggle/working/full_df_grouped.csv"
datos_csv_agrupados.to_csv(ruta_salida_grouped, index=False)
print(f"\nDataset con diagnósticos agrupados exportado como '{ruta_salida_grouped}'")

# Función para balancear datasets
def balancear_dataset(df, umbral_min):
    print("\nFrecuencia inicial de diagnósticos")
    distribucion_diagnosticos = df['diagnostico'].value_counts()
    print(distribucion_diagnosticos)
    
    subrepresentadas = distribucion_diagnosticos[distribucion_diagnosticos < umbral_min].index
    dominantes = distribucion_diagnosticos[distribucion_diagnosticos >= umbral_min].index
    
    print("\nCategorías subrepresentadas:")
    print(subrepresentadas)
    
    print("\nCategorías dominantes:")
    print(dominantes)
    
    df_sub = df[df['diagnostico'].isin(subrepresentadas)]
    df_dom = df[df['diagnostico'].isin(dominantes)]
    
    df_sub_resampled = resample(df_sub, replace=True, n_samples=len(df_dom), random_state=42)
    
    df_balanceado = pd.concat([df_dom, df_sub_resampled])
    
    print("\nDistribución después del balanceo")
    print(df_balanceado['diagnostico'].value_counts())
    
    return df_balanceado

# Balanceo de Data.xlsx
print("\nBalanceando Data.xlsx")
umbral_min_datos = 100
datos_balanceados = balancear_dataset(datos_excel, umbral_min_datos)

ruta_datos_balanceados = "/kaggle/working/data_balanced.xlsx"
datos_balanceados.to_excel(ruta_datos_balanceados, index=False)
print(f"\nDataset balanceado para Data.xlsx exportado como '{ruta_datos_balanceados}'")

# Balanceo de Full_df_cleaned.csv
print("\nBalanceando Full_df_cleaned.csv")
umbral_min_full_df = 200
full_df_balanceado = balancear_dataset(datos_csv, umbral_min_full_df)

ruta_full_df_balanceado = "/kaggle/working/full_df_balanced.csv"
full_df_balanceado.to_csv(ruta_full_df_balanceado, index=False)
print(f"\nDataset balanceado para Full_df_cleaned.csv exportado como '{ruta_full_df_balanceado}'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ruta_archivo = "/kaggle/working/full_df_balanced.csv"
datos = pd.read_csv(ruta_archivo)

print("\n=== Estadísticas descriptivas para las etiquetas binarias ===")
etiquetas = ['N', 'D', 'G', 'C', 'A', 'H', 'M', 'O']
estadisticas = datos[etiquetas].describe()
print(estadisticas)

print("\n=== Proporción de etiquetas activas (1) por columna ===")
for etiqueta in etiquetas:
    proporcion = datos[etiqueta].mean() * 100
    print(f"{etiqueta}: {proporcion:.2f}%")

print("\n=== Correlación entre etiquetas ===")
matriz_correlacion = datos[etiquetas].corr()
print(matriz_correlacion)

plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de Correlación entre Etiquetas Binarias")
plt.show()

```
